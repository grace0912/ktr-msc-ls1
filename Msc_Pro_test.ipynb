```python
from tkinter import*

def login_open(): # creation fonction login
    
    login_open=Tk()
    login_open.title("Data Protection :")
    login_open.geometry("400x300")
    login_open.minsize(400,300)
    login_open.maxsize(400,300)
    login_open.config(background="#007082")

    lblnbre10=Label(login_open, text = "Entrez User") #creation label User
    lblnbre10.place (x=50, y=50)      #placer le champ de saisi dans la fenetre
    entrynbre10=Entry(login_open)            #creation du premier point saisi
    entrynbre10.place (x=200, y=50)   #placer le champ de saisi dans la fenetre

    lblnbre20=Label(login_open, text = "Entrez PassWord") #creation label Password
    lblnbre20.place (x=50, y=100)      #placer le champ de saisi dans la fenetre
    entrynbre20=Entry(login_open)            #creation du premier point saisi
    entrynbre20.place (x=200, y=100)   #placer le champ de saisi dans la fenetre

    sign_up=Button(login_open, text="Login") # creation boutton sign_up
    sign_up.place(x=200, y=150)


    login_open.mainloop()


def library_interface(): # creation fonction library_interface pour ajouter new business
    library_interface=Tk()
    library_interface.title("ADD NEW BUSINESS CARDS :")
    library_interface.geometry("400x300")
    library_interface.minsize(400,300)
    library_interface.maxsize(400,300)
    library_interface.config(background="#007082")

    lblnbre11=Label(library_interface, text = "Name ") #creation label Nom
    lblnbre11.place (x=50, y=50)      #placer le champ de saisi dans la fenetre
    entrynbre11=Entry(library_interface)            #creation du premier point saisi
    entrynbre11.place (x=200, y=50)   #placer le champ de saisi dans la fenetre


    lblnbre21=Label(library_interface, text = "Company name") #creation label Nom Entreprise
    lblnbre21.place (x=50, y=100)      #placer le champ de saisi dans la fenetre
    entrynbre21=Entry(library_interface)             #creation du premier point saisi
    entrynbre21.place (x=200, y=100)   #placer le champ de saisi dans la fenetre

    lblnbre31=Label(library_interface, text = "Email address") #creation label Adresse mail
    lblnbre31.place (x=50, y=150)      #placer le champ de saisi dans la fenetre
    entrynbre31=Entry(library_interface)             #creation du premier point saisi
    entrynbre31.place (x=200, y=150)   #placer le champ de saisi dans la fenetre


    lblnbre41=Label(library_interface, text = "Phone Number") #creation label Numero phone
    lblnbre41.place (x=50, y=200)      #placer le champ de saisi dans la fenetre
    entrynbre41=Entry(library_interface)             #creation du premier point saisi
    entrynbre41.place (x=200, y=200)   #placer le champ de saisi dans la fenetre

    Ajouter=Button(library_interface, text="Ajouter la Saisie") # creation boutton ajouter un new business
    Ajouter.place(x=200, y=250)


    library_interface.mainloop()


Menu_Gen=Tk() # creation menu general pour faire le choix entre Login User et Library interface

lp1=Label(Menu_Gen, text="Menu General :", font="arial 18 italic", bg='white')
lp1.pack()

Menu_Gen.title("Menu General:") 
Menu_Gen.geometry("500x300")
Menu_Gen.config(background="#007082")

Menu_frame=Frame(Menu_Gen, width=310, height=135, bg="black") # creer frame pour loger les deux boutons

#creation du bouton login user
login_bouton= Button(Menu_frame, text="Login", font=("Arial",20),bg="white",fg="black", width=8, height=3, command=login_open)
login_bouton.grid(row=0,column=0)

#creation du bouton Library interface
library_bouton= Button(Menu_frame, text="Library interface", font=("Arial",20),bg="white",fg="black", width=12, height=3, command=library_interface)
library_bouton.grid(row=0,column=2)

Menu_frame.pack(expand=YES) # pour laisser le bouton au centre              
fen.mainloop()
```
