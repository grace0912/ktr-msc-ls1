```python
from tkinter import*
import os
from tkinter.messagebox import showerror, showinfo
#from listedesInscrit import listeInscrit

#creation fonction login:

def login_code() :
    nom_login = entrynbre10.get()
    pass_login = entrynbre20.get()
    File_login_name = os.listdir()
    if str(nom_login) + ".txt" in File_login_name :
        File1 = open(str(nom_login)+".txt", "r")
        liste_info_login = File1.read().split(":")
        File1.close()
        if pass_login == liste_info_login[1] :
            print("Message 1 : Bienvenu, votre nom et votre mot de passe sont juste : ")
            library_interface()
        else :
            print("Erreur 2 : le mot de passe est faux")
    else :
        print("Erreur 1: l'identifient n'existe pas")

    
# fonction Inscription_code:

def inscription_code() :
    nom_inscription = entrynbre101.get()
    pass_inscription = entrynbre201.get()
    File_inscription_name = os.listdir()
    if str(nom_inscription) + ".txt" in File_inscription_name :
        print("Erreur 3 : Le nom d'utilisateur est déjà pris")
    else :
        File = open(str(nom_inscription)+".txt", "w")
        File.write(str(nom_inscription)+":"+str(pass_inscription))
        File.close()
        print("Message 2 : Votre compte a bien été crée avec succès !")
        
# creation fonction window login:

def login_open(): 
    
    login_open=Tk()
    login_open.title("Data Protection :")
    login_open.geometry("400x300")
    login_open.minsize(400,300)
    login_open.maxsize(400,300)
    login_open.config(background="#007082")

    lblnbre10=Label(login_open, text = "Nom d'utilisateur") #creation label User
    lblnbre10.place (x=50, y=50)      #placer le champ de saisi dans la fenetre
    global entrynbre10
    entrynbre10=Entry(login_open)            #creation du premier point saisi
    entrynbre10.place (x=200, y=50)   #placer le champ de saisi dans la fenetre

    lblnbre20=Label(login_open, text = "Mot de passe") #creation label Password
    lblnbre20.place (x=50, y=100)      #placer le champ de saisi dans la fenetre
    global entrynbre20
    entrynbre20=Entry(login_open)            #creation du premier point saisi
    entrynbre20.place (x=200, y=100)   #placer le champ de saisi dans la fenetre

    sign_up=Button(login_open, text="se connecter", command=login_code) # creation boutton sign_up
    #sign_up=Button(login_open,text="se connecter",command=library_interface)
    sign_up.place(x=200, y=150)

    login_open.mainloop()
    
    
# creation fonction d'inscription :
def inscription_open(): 
    
    inscription_open=Tk()
    
    lp2=Label(inscription_open, text="Creer vos identifiants :", font="arial 18 italic", bg='white')
    lp2.pack()
    inscription_open.title("Creation compte:")
    inscription_open.geometry("400x300")
    inscription_open.minsize(400,300)
    inscription_open.maxsize(400,300)
    inscription_open.config(background="#007082")

    lblnbre101=Label(inscription_open, text = "Nom d'utilisateur") #creation label User
    lblnbre101.place (x=50, y=50)      #placer le champ de saisi dans la fenetre
    global entrynbre101
    entrynbre101=Entry(inscription_open)            #creation du premier point saisi
    entrynbre101.place (x=200, y=50)                #placer le champ de saisi dans la fenetre

    lblnbre201=Label(inscription_open, text = "Mot de passe") #creation label Password
    lblnbre201.place (x=50, y=100)      #placer le champ de saisi dans la fenetre
    global entrynbre201
    entrynbre201=Entry(inscription_open)            #creation du premier point saisi
    entrynbre201.place (x=200, y=100)   #placer le champ de saisi dans la fenetre

    sign_up=Button(inscription_open, text="Creer le compte", command=inscription_code) # creation boutton sign_up
    sign_up.place(x=200, y=150)

    inscription_open.mainloop()

    
    
    #creation class personnage à partir de son Nom, company,mail,phone:
class Personnage():
    def __init__(self,lblnbre11,lblnbre21,lblnbre31,lblnbre41):
        self.lblnbre11=lblnbre11
        self.lblnbre21=lblnbre21
        self.lblnbre31=lblnbre31
        self.lblnbre41=lblnbre41
        #init.mainloop()  
        
    def __eq__(self,other): # deux users n'auront pas les mêmes infos 
        return(self.lblnbre11== other.lblnbre11 and self.lblnbre31 == other.lblnbre31 and self.lblnbre41==other.lblnbre41)


def appartient(liste,val): # une fonction qui nous permet de savoir l'utilisateur a déjà été ajouté ou pas
    for i in range(len(liste)): # parcourir la liste
        if liste[i].__eq__(val):
            return 1 # vrai
    return 0  
    #appartient.mainloop()
      
def valider():
    global listePersonne
    #listePersonne=[]
    if entrynbre11.get() and entrynbre21.get() and entrynbre31.get() and entrynbre41.get() :
        pn=Personnage(entrynbre11.get(), entrynbre21.get(), entrynbre31.get(), entrynbre41.get()) # creation personnage
        if appartient(listePersonne, pn):
            showerror (title="Formulaire invalide!", message="cet utilisateur existe déjà.")
        else :  
            listePersonne.append(pn)
            showinfo (title="Validation reussie", message="{} a bien été ajouté".format(entrynbre11.get()))
    else :
        showerror (title="Formulaire invalide", message="Renseigner tous les champs")

    #valider.mainloop()         
                    
def reinitialiser(): # effacer tout ce qui se trouve dans les champs
    entrynbre11.delete(0,END)
    entrynbre21.delete(0,END)
    entrynbre31.delete(0,END)
    entrynbre41.delete(0,END)

    listePersonne=[]

    #reinitialiser.mainloop() 

def listedesinscrits(): # fonction liste des inscrits
    global listePersonne
    listedesinscrits=Tk()
    listedesinscrits.title("Liste des utilisateurs inscrits :")
    listedesinscrits.geometry("400x450")
    listedesinscrits.config(background="#007082")
    
    resultat=Label(listedesinscrits, text = "Liste des utilisateurs ajoutés : ")
    entryresultat=Entry(listedesinscrits)
    entryresultat.place(x=50,y=50)
    lblnbre11=Label(listedesinscrits, text = "Name")
    lblnbre21=Label(listedesinscrits, text = "Campany")
    lblnbre31=Label(listedesinscrits, text = "Email")
    lblnbre41=Label(listedesinscrits, text = "Telephone ")
    status=Label(listedesinscrits, text = "Aucun utilisateur ajouté")
    
    #listedesinscrits.grid(row=0,column=0)
    resultat.grid(row=0,column=0,columnspan=3)
    lblnbre11.grid(row=1,column=1,padx=10,pady=10)
    lblnbre21.grid(row=1,column=2,padx=10,pady=10)
    lblnbre31.grid(row=1,column=3,padx=10,pady=10)
    lblnbre41.grid(row=1,column=4,padx=10,pady=10)
    status.grid(row=2,column=0,columnspan=3)
    
    sign_upu=Button(listedesinscrits,text="se deconnecter",command=login_open)
    sign_upu.place(x=200, y=150)
    
    if listePersonne:
        r=2
        for p in listePersonne:
            no=Label(listedesinscrits, text=p.lblnbre11)
            n=Label(listedesinscrits, text=p.lblnbre21)
            e=Label(listedesinscrits, text=p.lblnbre31)
            t=Label(listedesinscrits, text=p.lblnbre41)
            
            no.grid(row=r, column=1)
            n.grid(row=r, column=2)
            e.grid(row=r, column=3)
            t.grid(row=r, column=4)
            listedesinscrits.create_line(9,55,355,55,width=1,fill='white')
            
            r+=1
            
            status.configure(text="{} ajouter pour le moment".format(len(listePersonne)))
            status.grid(row=r,column=0,columnspan=3,pady=2)
    
    listedesinscrits.mainloop()
    
    
  #creation fonction liste des inscrits       
                    
def library_interface(): # creation fonction library_interface pour ajouter new business
    library_interface=Tk()
    library_interface.title("ADD NEW BUSINESS CARDS :")
    library_interface.geometry("400x450")
    library_interface.minsize(400,450)
    library_interface.maxsize(400,450)
    library_interface.config(background="#007082")

    lblnbre11=Label(library_interface, text = "Name ") #creation label Nom
    lblnbre11.place (x=50, y=50)      #placer le champ de saisi dans la fenetre
    global entrynbre11
    entrynbre11=Entry(library_interface)            #creation du premier point saisi
    entrynbre11.place (x=200, y=50)   #placer le champ de saisi dans la fenetre


    lblnbre21=Label(library_interface, text = "Company") #creation label Nom Entreprise
    lblnbre21.place (x=50, y=100)      #placer le champ de saisi dans la fenetre
    global entrynbre21
    entrynbre21=Entry(library_interface)             #creation du premier point saisi
    entrynbre21.place (x=200, y=100)   #placer le champ de saisi dans la fenetre

    lblnbre31=Label(library_interface, text = "Email") #creation label Adresse mail
    lblnbre31.place (x=50, y=150)      #placer le champ de saisi dans la fenetre
    global entrynbre31
    entrynbre31=Entry(library_interface)             #creation du premier point saisi
    entrynbre31.place (x=200, y=150)   #placer le champ de saisi dans la fenetre


    lblnbre41=Label(library_interface, text = "Phone") #creation label Numero phone
    lblnbre41.place (x=50, y=200)      #placer le champ de saisi dans la fenetre
    global entrynbre41
    entrynbre41=Entry(library_interface)             #creation du premier point saisi
    entrynbre41.place (x=200, y=200)   #placer le champ de saisi dans la fenetre

    Ajouter=Button(library_interface, text="valider", command=valider) # creation boutton ajouter un new business
    Ajouter.place(x=200, y=250)
    
    Ajouter1=Button(library_interface, text="Reinitialiser", command=reinitialiser) # creation boutton renitialiser
    Ajouter1.place(x=200, y=300)
    
    Ajouter2=Button(library_interface, text="Voir la liste",command=listedesinscrits) #,command=lambda:listedesInscrits(fenetre,listePersonne)) # creation boutton voir la liste des utilisateurs
    Ajouter2.place(x=200, y=350)

    library_interface.mainloop()
        
    

  # creation menu general pour faire le choix entre Login User et Library interface  
    
Menu_Gen=Tk() 

lp1=Label(Menu_Gen, text="Menu General :", font="arial 20 italic", bg='white')
lp1.pack()

Menu_Gen.title("Menu General : ") 
Menu_Gen.geometry("500x300")# dimensionné avec le point geometry 500*300
Menu_Gen.config(background="#007082")

Menu_frame=Frame(Menu_Gen, width=310, height=135, bg="black") # creer frame pour loger les deux boutons

#creation du bouton login user
login_bouton= Button(Menu_frame, text="Login", font=("Arial",20),bg="white",fg="black", width=8, height=3, command=login_open)
login_bouton.grid(row=0,column=0)

#creation du bouton inscription ( entrez vos identifiants d'inscription)
library_bouton= Button(Menu_frame, text="Inscription", font=("Arial",20),bg="white",fg="black", width=12, height=3, command=inscription_open)
library_bouton.grid(row=0,column=2)

Menu_frame.pack(expand=YES) # pour laisser le bouton au centre   

Menu_Gen.mainloop()
```
